```python
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
```

```python
print('Dimensions:', arr.ndim)
print('Shape', arr.shape)
print('Size', arr.size)
print('Type', arr.dtype)
```

```python
# Special methods to create new array 
```

```python
zeros = np.zeros((3,5),dtype = int)
print(zeros)
print(zeros.dtype)
```

```python
ones = np.ones((3,3,3),dtype = int)
print(ones)
print(ones.dtype)
```

```python
empty = np.empty((2,3))
print(empty)
```

```python
#Arange
print(np.arange(10))
```

```python
print(np.arange(5,10))
```

```python
print(np.arange(0,20,2))
# from 0 to 19 with step 2
```

```python
# numpy.linspace

# linspace is used to generate evenly spaced numbers over a specified interval.

# numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)
arr = np.linspace(0, 10, 5)
print(arr)

```

```python
print(np.linspace(2,8,4))
      
```
