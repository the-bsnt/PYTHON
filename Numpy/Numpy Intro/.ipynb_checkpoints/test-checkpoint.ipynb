```python
import numpy as np
vector = np.array([1,2,3,4,5])
matrix = np.array([[1,2,3],[4,5,6]])
print(vector)
print(matrix)
```

```python
vector = np.array([1,2,3,4,5.5])
print(vector)
vector = np.array([1,2,3,4,5,'A'])
print(vector)

# the data in numpy array have to be of same type, if not the data is type casted to compactible datatype. like if one char present convert all data to char 
# or str 
```

```python
# Numpy array -> Fixed size ( NOt resizable)
arr = np.array([1,2,3])
new_arr = np.append(arr, 4)
print(arr)
print(new_arr)
```

```python
# Functions like np.array, np.concatenate, np.insert and np.delete return a new array , they dont resize the original one 
```

```python
# Array Basics in NUMpy
arr = np.array([1,2,3,4,5])
print(arr[3])
```

```python
# numpy arrays are zero indexed and mutable also 
```

```python
arr[3] = 56
print(arr[3])
print(arr)
```

```python
print(arr[1:3])
print(arr[::2]) #step of 2 
print(arr[3::])
```
