```python
import numpy as np
arr = np.arange(3)
arr = arr[np.newaxis, :]
print(arr)
print(arr.shape)
```

```python
arr2= np.arange(3)
arr2 = arr2[:, np.newaxis]
print(arr2)
print(arr2.shape)
```

```python
arr3 = np.arange(3)
expanded = np.expand_dims(arr3,axis= 1)
print(expanded)
print(expanded.shape)
```

```python
#Slicing

arra = np.array([[0,1],[2,3],[30,40]])
print(arra[arra % 2 == 0 ]) # you can give conditions to filter out the array
```

```python
print(arra[arra > 20])
```

```python
arr = np.arange(20)
print(arr[(arr % 2 == 0) & (arr > 10)])
```

```python
conditions =(arr % 2 == 0) & (arr > 10)
print(conditions)
```

```python
# For indexis that met the conditon 
matrix = np.array([[3,5,6],[7,6,9]])
print(matrix)
print(np.nonzero(matrix % 2 != 0))
print(np.nonzero(matrix== 0))
```
