```python
import numpy as np

arr = np.array([2,6,3,7,37,74,9])
print(np.sort(arr))
```

```python
arr.sort()
print(arr)


# np.sort(arr) → Function → Returns new array (original unchanged)

# arr.sort() → Method → Modifies array in-place (original changed)
```

```python
# Concatinating two arrays

a = np.array([3,46,23,65])
b = np.array([4, 56,75,80])
c = np.array([4, 56,75,80])

print(np.concatenate([a,b,c])) # we have to pass a,b in form of iterable either list or tuple
```

```python
print(np.sort(np.concatenate([a,b,c])))
```

```python
# if array are of different dimensions
x = np.array([[1,2,3],[4,5,6]])
y = np.array([[7,8,9]])
print(np.concatenate([x,y]))
```

```python
# Using axis parameter explicitly
print(np.concatenate([x, y], axis=0))  # Same as default

# Using vstack (vertical stack)
print(np.vstack([x, y]))

# Using r_ (concatenation shortcut)
print(np.r_[x, y])
```

```python
z = np.array([[7,8,9],[0,0,0]])
# Using axis parameter explicitly axis =1
print(np.concatenate([x, z], axis =1)) 
```

```python
# Reshaping the arrays 

# 2D
arr = np.arange(12)
print(arr)
print(arr.reshape(6,2))
print(arr.reshape(2,6))
print(arr.reshape(2,2,3))

```

```python
# 3D
arr = np.arange(27)
print(arr)
x= arr.reshape(3,3,3)
print(x)
print(x.reshape(-1)) # converts back to 1 D array
# or
print(x.reshape(27))


```
