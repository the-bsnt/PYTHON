```python
import numpy as np
import matplotlib.pyplot as plt

points = np.array([[1, 2, 3],  # Point 1: (1, 2, 3)
                   [4, 5, 6]]) # Point 2: (4, 5, 6)
```

```python
points_2d = np.array([[1, 4],   # Point 1: (1, 4)
                      [2, 5],   # Point 2: (2, 5)
                      [3, 6]])  # Point 3: (3, 6)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

# Plot as two points
ax.scatter(matrix[:, 0], matrix[:, 1], matrix[:, 2], c='r', s=100)

# Plot as vectors from origin
for i in range(2):
    ax.quiver(0, 0, 0, matrix[i, 0], matrix[i, 1], matrix[i, 2], 
              arrow_length_ratio=0.1, color=['r', 'b'][i])
```
